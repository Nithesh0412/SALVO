```python
#first let's import the necessary lib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import  train_test_split
```

```python
X=df.drop(columns='label',axis=1)
Y=df['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
#So now lets split the data into train and test data
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

```python
#we need to standardize the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train_std=scaler.fit_transform(X_train)
X_test_std=scaler.transform(X_test)
```

```python
#SO 20percent of the data is used for testing purpose and the remaining 80percent of the data is used for training purpose
```

```python
#import tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
#this set_seed is used so that the accuracy value don't change each time when we run the program
```

```python
#now lets setup the first layer for our Neural network
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
])

#this dense means lets say we have four neuron in the hidden layer and three neuron in the input layer so if we see one of the neuron in the hidden layer three input would be connected
#the flatten converts the data into an single dimension array
#relu MEANS RECTIFIED LINEAR UNIT
```

```python
#compiling neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
              )
#so here this 'sparse_categorical_crossentropy' is used since we have the label as either 1 or 0
```

```python
history=model.fit(X_train_std,Y_train,validation_split=0.1,epochs=10)
#epochs is like how many times the model has to go through the data
```

```python
#so heere the loss and accurate value are inversely proportional
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc='upper right')
```

```python
#now the data will be tested and will be compared with the true dataa by evaluate
loss,accuracy=model.evaluate(X_test_std,Y_test)
print(accuracy)
```

```python
Y_pred=model.predict(X_test_std)
print(X_test_std.shape)
print(X_test_std[0])
```
